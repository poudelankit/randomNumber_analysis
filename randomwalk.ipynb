**Random Walk**

```python
'''Random Walk
1. Start the program
2. Set step size =1
3. initialize displacement to 0.0
4. you write down parts of code to generate RN.
5. start loop (1 to N)
    let RN if (RN > 0.5) > displacement = displacement +1
    else displacement =  displacement -1
6. find the cumulative displacement for N steps'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random 
random.random()
```

```python
no_of_steps = 1000
step_size = 1
displacements_made = [0]
displacement = 0
for i in range(1,no_of_steps):
    random_num = random.random()
    if random_num > 0.5:
        displacement=1
    else:
        displacement=-1
    displacements_made.append(displacement)
cummulative_sum_displacement = np.array(displacements_made).cumsum()
start = cummulative_sum_displacement[:1]
stop = cummulative_sum_displacement[-1:]
fig = plt.figure(figsize=(8,4),dpi=200)
ax = fig.add_subplot(111)
ax.scatter(np.arange(no_of_steps),cummulative_sum_displacement,c='blue',s=0.05,alpha=0.5)
ax.plot(cummulative_sum_displacement,c='blue',alpha=0.8,lw=0.2,ls='--')
ax.plot(0, start, c='red', marker='+')
ax.plot(no_of_steps,stop,c='black',marker='o')
plt.title('1D Random Walk')
plt.tight_layout(pad=0)
plt.show()
```

```python
dims = 1
step_n = 1000
step_set = [-1,1]
origin = np.zeros((1, dims))
step_shape = (step_n,dims)
steps = np.random.choice(a=step_set,size=step_shape)
path=np.concatenate([origin,steps]).cumsum()
start = path[:1]
stop = path[-1:]
#plot the path
fig = plt.figure(figsize=(8,4),dpi=200)
ax = fig.add_subplot(111)
ax.scatter(np.arange(step_n+1),path,c='blue',s=0.05,alpha=0.5)
ax.plot(path,c='blue',alpha=0.8,lw=0.2,ls='--')
ax.plot(0, start, c='red', marker='+')
ax.plot(step_n,stop,c='black',marker='o')
plt.title('1D Random Walk')
plt.tight_layout(pad=0)
plt.show()
```
