**Random Walk**

```python
'''Random Walk
1. Start the program
2. Set step size =1
3. initialize displacement to 0.0
4. you write down parts of code to generate RN.
5. start loop (1 to N)
    let RN if (RN > 0.5) > displacement = displacement +1
    else displacement =  displacement -1
6. find the cumulative displacement for N steps'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random 
random.random()
```

```python
no_of_steps = 1000
step_size = 1
displacements_made = [0]
displacement = 0
for i in range(no_of_steps):
    random_num = random.random()
    if random_num > 0.5:
        displacement=1
    else:
        displacement=-1
    displacements_made.append(displacement)
cummulative_sum_displacement = np.array(displacements_made).cumsum()
start = cummulative_sum_displacement[:1]
stop = cummulative_sum_displacement[-1:]
fig = plt.figure(figsize=(8,4),dpi=200)
ax = fig.add_subplot(111)
ax.scatter(np.arange(no_of_steps+1),cummulative_sum_displacement,c='blue',s=0.05,alpha=0.5)
ax.plot(cummulative_sum_displacement,c='blue',alpha=0.8,lw=0.2,ls='--')
ax.plot(0, start, c='red', marker='+')
ax.plot(no_of_steps,stop,c='black',marker='o')
plt.title('1D Random Walk')
plt.tight_layout(pad=0)
plt.show()
```

*The above graph show the random walk in 1D space, where we consider displacement in either positive or negetive axis and the displacement is by 1 step size. The red (+) sign show the initial position where as black (o) show the position at the Nth iteration.* 

*If 'd' is the displacement made by drunkard each time we conduct the experiment for N steps. <br> If we conducted this experiment for multiple times (say m times tends to inf) then the averge distance travelled is equal to 0 as each step has equal probability of being +1 or -1*

**Lets find the average of square of displacements**

```python
np.sqrt(sum([x**2 for x in displacements_made]))
```

```python
np.sqrt(no_of_steps)
```

*The average of the square of the distance is equal to N. If we take the square root of this equation, we realize that:
sqrt(d^2)=sqrt(N)
Since sqrt(d^2) is something like the average positive distance away from 0 after N steps (technically, it's called the "root-mean-squared" distance), we expect that after N steps, the drunken man will be roughly sqrt(N) steps away from where he started. For example:  for 25 steps, we expect the drunken man to have moved roughly 5 total spaces from 0 in either direction. Of course sometimes it will move more and sometimes fewer total spaces, but 5 is roughly what we might expect.*
